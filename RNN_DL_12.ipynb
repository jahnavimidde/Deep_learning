```python
from keras.models import Sequential
from keras.layers import SimpleRNN, Dense
import numpy as np

# Vocabulary (characters)
vocab = ['d', 'e', 'p', '<stop>']
char2idx = {c: i for i, c in enumerate(vocab)}
idx2char = {i: c for c, i in char2idx.items()}

# Sequence: "dep"
sequence = ['d', 'e', 'p']
target = ['e', 'p', '<stop>']

# One-hot encoding
X = np.eye(len(vocab))[[char2idx[c] for c in sequence]].reshape(len(sequence), 1, len(vocab))
y = np.eye(len(vocab))[[char2idx[c] for c in target]]

# Model
model = Sequential([
    SimpleRNN(8, input_shape=(1, len(vocab))),
    Dense(len(vocab), activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam')
model.fit(X, y, epochs=200, verbose=0)

# Predict next char after 'p'
test = np.eye(len(vocab))[char2idx['p']].reshape(1,1,len(vocab))
pred = model.predict(test)
print("Next char:", idx2char[np.argmax(pred)])
```

```python
# Vocabulary (words)
vocab = ['deep', 'learning', 'is', '<stop>']
word2idx = {w: i for i, w in enumerate(vocab)}
idx2word = {i: w for w, i in word2idx.items()}

sequence = ['deep', 'learning', 'is']
target = ['learning', 'is', '<stop>']

# One-hot encoding
X = np.eye(len(vocab))[[word2idx[w] for w in sequence]].reshape(len(sequence), 1, len(vocab))
y = np.eye(len(vocab))[[word2idx[w] for w in target]]

# Model
model = Sequential([
    SimpleRNN(8, input_shape=(1, len(vocab))),
    Dense(len(vocab), activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam')
model.fit(X, y, epochs=300, verbose=0)

# Predict next word after 'is'
test = np.eye(len(vocab))[word2idx['is']].reshape(1,1,len(vocab))
pred = model.predict(test)
print("Next word:", idx2word[np.argmax(pred)])
```

```python
# Sentence vocabulary
vocab = [
    'deep learning is fun',
    'machine learning is powerful',
    'AI is the future',
    '<stop>'
]

sent2idx = {s: i for i, s in enumerate(vocab)}
idx2sent = {i: s for s, i in sent2idx.items()}

# Input sequence (sentences)
sequence = [
    'deep learning is fun',
    'machine learning is powerful',
    'AI is the future'
]

target = [
    'machine learning is powerful',
    'AI is the future',
    '<stop>'
]

# One-hot encoding
X = np.eye(len(vocab))[[sent2idx[s] for s in sequence]].reshape(len(sequence), 1, len(vocab))
y = np.eye(len(vocab))[[sent2idx[s] for s in target]]

# Model
model = Sequential([
    SimpleRNN(16, input_shape=(1, len(vocab))),
    Dense(len(vocab), activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam')
model.fit(X, y, epochs=500, verbose=0)

# Predict next sentence
test = np.eye(len(vocab))[sent2idx['AI is the future']].reshape(1,1,len(vocab))
pred = model.predict(test)
print("Next sentence:", idx2sent[np.argmax(pred)])
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
import numpy as np

# Vocabulary
vocab = ['deep', 'learning', 'is', 'very', 'powerful', '<stop>']
word2idx = {w:i for i,w in enumerate(vocab)}
idx2word = {i:w for w,i in word2idx.items()}

# Sequence
sequence = ['deep', 'learning', 'is', 'very']
target   = ['learning', 'is', 'very', 'powerful']

# One-hot encoding
X = np.eye(len(vocab))[[word2idx[w] for w in sequence]]
y = np.eye(len(vocab))[[word2idx[w] for w in target]]

# Reshape for LSTM → (samples, timesteps, features)
X = X.reshape(1, len(sequence), len(vocab))
y = y.reshape(1, len(target), len(vocab))

# Model
model = Sequential([
    LSTM(32, return_sequences=True, input_shape=(len(sequence), len(vocab))),
    Dense(len(vocab), activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam')

# Train
model.fit(X, y, epochs=300, verbose=0)

# Predict next word after sequence
pred = model.predict(X)
print("Next word:", idx2word[np.argmax(pred[0, -1])])
```

```python
from keras.models import Sequential
from keras.layers import GRU, Dense
import numpy as np

# Same vocabulary
vocab = ['deep', 'learning', 'is', 'very', 'powerful', '<stop>']
word2idx = {w:i for i,w in enumerate(vocab)}
idx2word = {i:w for w,i in word2idx.items()}

sequence = ['deep', 'learning', 'is', 'very']
target   = ['learning', 'is', 'very', 'powerful']

# One-hot encoding
X = np.eye(len(vocab))[[word2idx[w] for w in sequence]]
y = np.eye(len(vocab))[[word2idx[w] for w in target]]

X = X.reshape(1, len(sequence), len(vocab))
y = y.reshape(1, len(target), len(vocab))

# Model
model = Sequential([
    GRU(32, return_sequences=True, input_shape=(len(sequence), len(vocab))),
    Dense(len(vocab), activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam')

# Train
model.fit(X, y, epochs=300, verbose=0)

# Prediction
pred = model.predict(X)
print("Next word:", idx2word[np.argmax(pred[0, -1])])
```
